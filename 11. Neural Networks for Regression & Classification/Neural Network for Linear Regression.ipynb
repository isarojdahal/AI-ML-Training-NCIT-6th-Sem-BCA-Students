```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# Sample dataset: [size], [price]
X = np.array([[500], [1000], [1500], [2000], [2500], [3000], [3500]], dtype=float)
y = np.array([[50], [100], [150], [200], [250], [300], [350]], dtype=float)  # in lakhs

# Normalize data
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

# Split train/test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

# Neural Network model
model = Sequential([
    Input(shape=(1,)), # Use Input layer
    Dense(10, activation='relu'),
    Dense(5, activation='relu'),
    Dense(1)   # regression output (no activation)
])

model.compile(optimizer='adam', loss='mse')

# Train model
model.fit(X_train, y_train, epochs=20) #fourth param, verbose = 1 by default.

# Predict
predicted = scaler_y.inverse_transform(model.predict(X_test)) # transform: [[0.52], [0.74]]  в†’ [[152.3], [218.7]]  (lakhs)
print("Predicted Prices:", predicted.flatten()) # flatten : [[152.3], [218.7]] в†’ [152.3, 218.7]
```
