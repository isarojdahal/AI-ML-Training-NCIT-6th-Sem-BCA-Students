```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelBinarizer
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target.reshape(-1, 1)  # labels (0,1,2)

# One-hot encoding for classification
encoder = LabelBinarizer()
y_encoded = encoder.fit_transform(y)


# Normalize features
scaler = MinMaxScaler() #also we have StandardScaler
X_scaled = scaler.fit_transform(X)

# Split train/test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2,
                                                    random_state=42)

# Neural Network
model = Sequential([
    Dense(10, input_dim=4, activation='relu'),
    Dense(8, activation='relu'),
    Dense(3, activation='softmax')  # 3 classes → softmax
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
# Train
model.fit(X_train, y_train, epochs=100, verbose=0)

# Evaluate
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {acc*100:.2f}%")

```

```python

new_data = np.array([[5.1, 3.5, 1.4, 0.2]]) # Example data for a Setosa

new_data_scaled = scaler.transform(new_data)

# Make a prediction
new_prediction = model.predict(new_data_scaled)

# Decode the prediction
predicted_class_decoded = encoder.inverse_transform(new_prediction)[0]

# Display the predicted class
print(f"\nPredicted class for the new data: {predicted_class_decoded}")
```
