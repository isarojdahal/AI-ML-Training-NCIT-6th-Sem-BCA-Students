```python
doc = """ 
DevOps and MLOps both aim to automate and streamline the path to production, but they manage fundamentally different assets. DevOps focuses exclusively on software code and infrastructure, ensuring that applications can be built, tested, and deployed reliably. MLOps extends this philosophy to the machine learning domain, meaning it must orchestrate not just the code, but also massive datasets, hyperparameter configurations, and trained model weights to ensure reproducible predictive models.

This difference in assets completely changes how testing is handled. In a standard DevOps pipeline, Continuous Integration (CI) involves unit, integration, and end-to-end tests to guarantee the software logic is sound. In MLOps, CI is much broader; before you even test the code, you must validate incoming data schemas, check for anomalous values, and evaluate the trained model's predictive accuracy and fairness against holdout datasets to ensure it meets baseline performance metrics.

Deployment in these two disciplines also produces different artifacts. While DevOps typically deploys a compiled application, web server, or microservice, MLOps deploys a prediction service alongside specialized infrastructure. This operational environment often includes a model registry to track which specific version of a model's weights is actively serving users, and a feature store designed to supply the model with real-time, pre-calculated data points needed for live inference.

The most critical distinction emerges after deployment during the monitoring phase. A traditional software system managed by DevOps generally remains stable in its behavior unless the code or servers break. In contrast, an MLOps system will naturally degrade over time—even if the code never changes—because real-world behaviors evolve, causing "data drift." To combat this, MLOps introduces Continuous Training (CT), an automated loop that detects when a model's predictions become inaccurate and triggers a pipeline to retrain it on fresh data.

""" 
```

```python

from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)

chunked_texts = text_splitter.split_text(doc)
chunked_texts
```

```python
len(chunked_texts)
```

```python
from langchain_ollama import OllamaEmbeddings

embedding_model = OllamaEmbeddings(model="nomic-embed-text:latest")
```

```python
from langchain_core.vectorstores import InMemoryVectorStore

vector_store = InMemoryVectorStore(embedding_model)

vector_store.add_texts(chunked_texts)
```

## Workflow 2

```python
prompt = "what is AI?" 
```

```python
result = vector_store.similarity_search(query=prompt,k=1)
result
```

```python
final_prompt = f"""

{prompt}

{result[0].page_content}

Context : Summarize the response in few sentences in nepali language.

"""

final_prompt
```

```python
from langchain_ollama import ChatOllama

model = ChatOllama(model="llama3.2:latest")
res = model.invoke(final_prompt)
print(res.content)
```
