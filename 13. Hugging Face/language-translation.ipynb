```python

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# For Nepali to English translation, we'll use another general multilingual Opus-MT model.
tokenizer_mul_en = AutoTokenizer.from_pretrained("iamTangsang/MarianMT-Nepali-to-English-Translation-Final")
model_mul_en = AutoModelForSeq2SeqLM.from_pretrained("iamTangsang/MarianMT-Nepali-to-English-Translation-Final")

nepali_text_en = "नमस्ते, हजुरलाई आज कस्तो छ? मलाई आशा छ कि तपाईले हगिङ फेस मोडलहरू प्रयोग गरेर राम्रो समय बिताउनुहुनेछ!"

# Prepare input for the translation model
inputs_mul_en = tokenizer_mul_en(nepali_text_en, return_tensors="pt")
translated_ids_mul_en = model_mul_en.generate(inputs_mul_en.input_ids)
english_translation_ne = tokenizer_mul_en.decode(translated_ids_mul_en[0], skip_special_tokens=True)

print(f"Original Nepali: {nepali_text_en}")
print(f"Translated English: {english_translation_ne}")
```
