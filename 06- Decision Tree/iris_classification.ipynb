```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as pyplot 
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = load_iris()
print(iris.data)
```

```python
print(iris.data.shape) # gives number of rows and cols; eg: (150,4)
print(iris.feature_names) # feature name = column name
print(iris.target_names) # classes.
```

```python
x = iris.data 
y = iris.target
```

```python
#AI Train
model = DecisionTreeClassifier().fit(x,y)
```

```python
predictions = model.predict([[1.1,1.2,3.0,2.3]]) 
# 1.1 = sepal length, 1.2 = sepal width
# 3.0  = petal length , 2.3 = petal width

print(iris.target_names[predictions[0]])
```

```python
from sklearn.ensemble import RandomForestClassifier

model2 = RandomForestClassifier(n_estimators=200).fit(x,y)

predictions2 = model2.predict([[1.1,1.2,3.0,2.3]])

print(iris.target_names[predictions2[0]])

```
