```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'house_price':[245,312,279,308,199,219,405,324,319,255], #lakhs
    'square_foot':[1400,1600,1700,1875,1100,1550,2350,2450,1425,1700]
})
df
```

```python
import seaborn as sns

sns.scatterplot(data=df,x='square_foot',y='house_price')
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()

X=df[['square_foot']]
y=df['house_price']

model = reg.f # y = mx+c

print(model.coef_) #slope
print(model.intercept_) #c
 
```

```python
predicted_value = model.predict(X=[[1800]]) # 1800 sq. foot , how much price does it cost ?

print(predicted_value[0])
```

```python

import matplotlib.pyplot as plt 

sns.scatterplot(data=df,x='square_foot',y='house_price')
plt.plot(X,reg.predict(X),color="green")

```

```python
y_all_predicted = model.predict(X)

from sklearn.metrics import r2_score

accuracy = r2_score(y,y_all_predicted)

print("accuracy is :"+str(accuracy))
```

```python
from sklearn.metrics import mean_absolute_error

accuracy2 = mean_absolute_error(y,y_all_predicted)

print("MAE is "+str(accuracy2))
```
