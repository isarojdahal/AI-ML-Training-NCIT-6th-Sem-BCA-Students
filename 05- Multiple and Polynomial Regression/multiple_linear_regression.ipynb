```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'house_price':[245,312,279,308,199,219,405,324,319,255], #lakhs
    'square_foot':[1400,1600,1700,1875,1100,1550,2350,2450,1425,1700],
    'no_of_rooms':[8,10,11,13,6,9,15,16,9,10]
})
df
```

```python
import seaborn as sns

sns.scatterplot(data=df,x='square_foot',y='house_price')
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()

X=df[['square_foot','no_of_rooms']]
y=df['house_price']

model = reg.fit(X,y)

print(model.coef_) #slope
print(model.intercept_) #c
 
```

```python
predicted_value = model.predict(X=[[1600,3]]) # 1600 sq. foot , 3 rooms how much price does it cost ?

print(predicted_value[0])
```
